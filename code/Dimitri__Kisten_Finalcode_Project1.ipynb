#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def get_stddev(column):
    stddev = 0 #Made sure to set to 0 outside of function
    average = 0 #same for average
    column = [column]
    average = np.mean(column) #function to generate the average for each column
    diffsquared = 0
    sum_diffsquared = 0
    for x in column: #iterating through each value in the column
        diffsquared = (x-average)**2 #difference of individual mean and total average squared. 
        sum_diffsquared = diffsquared + sum_diffsquared #Summing the squared differences
    stddev = ((sum_diffsquared)/len(column))**(1/2)  #Taking the sum of squared differences dividing by n, and sqrt.
    print(np.sqrt(np.var(column)))
````

````python
def subplot_histograms(dataframe, columns, titles, xlabels):
    nrows = int(np.ceil(len(columns)/2)) #defines the number of rows for the output
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize = (14,14)) 
    # Figsize to enlarge and allow for more spacing and visualization of values
    ax = ax.ravel() #Turns matrix --> vector --> iterate through
    for a, column in enumerate(columns): 
        ax[a].hist(dataframe[column]) 
        ax[a].set_title(titles[a], fontsize = 18)
        ax[a].set_xlabel(xlabels[a])
    plt.tight_layout() #organizes histograms better
    plt.show()

      
````

#### Plot and interpret boxplots

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels
